Conditional  with loc and iloc

```python
#Conditional with loc 
import pandas as pd

Df = pd.DataFrame({
    "name":   ["Alice", "Bob", "Carol", "David" , "Bob"],
    "city":   ["New York", "Chicago", "Houston", "Seattle" , "Chicago"],
    "salary": [85000, 72000, 91000, 67000 , 72000]
}, index=[10, 20, 30, 40 , 50]) 

# Conditional selection using loc

Df.loc[(Df["salary"] > 70000) & (Df["salary"] < 90000)]

# OR → at least one must be true
Df.loc[(Df["city"] == "Chicago") | (Df["city"] == "Seattle")]

# NOT → negate a condition
Df.loc[~(Df["city"] == "New York")]

#Condition + select specific column

# salary of employees in Seattle only
Df.loc[Df["city"] == "Seattle", "salary"]

# name and salary where salary > 80000
Df.loc[Df["salary"] > 80000, ["name", "salary"]]


# Condition + update values

# give a raise to employees earning less than 75000
Df.loc[Df["salary"] < 75000, "salary"] += 5000

# update city for a specific name
Df.loc[Df["name"] == "Bob", "city"] = "New York"
```

Apply function on df

```python
# apply is used to apply a function to each element of a column or row in a DataFrame.

import pandas as pd
import numpy as np

df1 = pd.DataFrame({
    'Math':    [85, 92, 78, 90],
    'Science': [88, 79, 95, 83],
    'English': [72, 85, 80, 91]
}, index=['Alice', 'Bob', 'Charlie', 'David'])

print(df1)

# Apply along columns (axis=0) — stats per subject
print(df1.apply(np.mean, axis=0))

# Apply along rows (axis=1) — stats per student
df1['Average'] = df1.apply(np.mean, axis=1)
df1['Grade']   = df1['Average'].apply(lambda x: 'A' if x >= 85 else 'B')

print(df1)
```

```python
mn=df["Total Marks"].min()
max=df["Total Marks"].max()

df["Scaled Marks"]=df["Total Marks"].apply(lambda x : (x-mn)/(max-mn))
df
```

```python
# custom function 

def grade(marks) :
    if marks >= 270 :
        return "A+"
    elif marks >= 240 :
        return "A"
    elif marks >= 210 :
        return "B"
    else :
        return "C"

df["Grade"] = df["Total Marks"].apply(grade)  ## column wise 
df   
    
```

```python
def making_system(df) :
    a = df["Math"]+2
    b= df["English"]+2
    c= df["Physics"]+2
    return a+b+c

df["Raised Total Mark"] =df.apply(making_system , axis=1 ) # row wise 
df
```
