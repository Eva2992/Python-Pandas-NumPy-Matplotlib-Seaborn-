2D Line Plot 

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
hours =[2,3,4,1,5,3,7]
days=['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
plt.plot(days,hours) # x->days, y->hours
plt.xlabel('Days of the week')
plt.ylabel('Hours of sleep')
plt.title('Hours of sleep vs Days of the week')
```

```python
hours_1 =[2,3,4,1,5,3,7]
hours_2=[1,2,3,4,5,6,7]
plt.plot(days,hours_1, label='Person 1') 
plt.plot(days,hours_2, label='person_1')
plt.xlabel('Days of the week')
plt.ylabel('Hours of sleep')
```

#2D plot from pandas Data

```python
df=pd.read_csv('enroll_data.csv')
df
```

```python
plt.plot(df['Year'] , df['Programming'] , label="Programming")
plt.plot(df["Year"] , df["Digital Marketing"] , label="Digital Marketing")

plt.xlabel('Year')
plt.ylabel("Programming Interests")
plt.legend()
#plt.legend(loc='upper left')
```

```python
plt.plot(df['Year'] , df['Programming'] , label="Programming" , color='green' , linewidth=3 , linestyle="--"  , marker='o')
plt.plot(df["Year"] , df["Digital Marketing"] , label="Digital Marketing" , color='red' , linestyle="dotted" , marker="o" )

plt.xlabel('Year')
plt.ylabel("Programming Interests")
plt.legend()
```

```python
plt.plot(df['Year'] , df['Programming'] , label="Programming" , color='green' , linewidth=3 , linestyle="--"  , marker='o')
plt.plot(df["Year"] , df["Digital Marketing"] , label="Digital Marketing" , color='red' , linestyle="dotted" , marker="o" )

plt.xlabel('Year')
plt.ylabel("Programming Interests")
plt.legend()
plt.grid()
#plot.show()
# for more example of matplotlib visit https://matplotlib.org/stable/gallery/index.html
```

HistoGram

```python
# Histogram is used to show the distribution of a single variable. 
# It is a type of bar chart that represents the frequency of data points 
# within specified ranges (bins). Each bar in a histogram represents the count 
# of data points that fall within a particular bin, 

plt.hist(df["IQ_Score"]  , bins=6 , edgecolor='green')
plt.xlabel('IQ Score')
plt.ylabel('Frequency of students')
```

```python
plt.hist(df["Study_Hour"] , edgecolor='skyblue' , color='lightblue')
plt.xlabel('Study Hours')
plt.ylabel('Frequency of students')
```

```python
plt.hist(df["Study_Hour"] , bins=[0,1,2,3,4,5,6,7] , edgecolor='skyblue' , color='lightblue')
plt.xlabel('Study Hours')
plt.ylabel('Frequency of students')
```

Bar Chart 

```python
data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df=pd.DataFrame(data)
df
```

```python
# For more example of scatter plot visit https://matplotlib.org/stable/gallery/index.html

#Gender Wise 

grouped_by = df.groupby("Gender").size()
print(grouped_by)
print(grouped_by.index)
print(grouped_by.values)
```

```python
plt.bar(grouped_by.index ,grouped_by.values , color =["blue" , "pink"] , edgecolor='black')

plt.xlabel('Gender')
plt.ylabel('Number of students')
plt.title("Gender wise distribution of students")
```
