Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
#seaborn is a Python data visualization library based on matplotlib.
#  Seaborn is built on top of matplotlib and integrates closely with pandas data structures, 
# making it easy to create complex/multiple visualizations 
# as well as functions for visualizing distributions, relationships, and categorical data.
```

```python
student= pd.read_csv("./sns_data.csv")
student.head()
```

LinePlot using Seaborn

```python
sns.lineplot(data = student , x="week" , y="attendance_rate" , errorbar=None)
```

```python
sns.lineplot(data=student , x="week" , y="attendance_rate" , hue="gender" , errorbar=None)
```

```python
sns.relplot(kind="line" , data =student , x="week" , y="test_score" , hue="gender" , errorbar=None)

# replot is figure level plot , can generate multiple subplot
```

```python
sns.relplot(kind="line" , data =student , x="week" , y="test_score" , hue="class_level" , errorbar=None)
```

```python
sns.scatterplot(data=student , x='study_hours' , y='test_score' )
```

```python
sns.scatterplot(data=student , x='study_hours' , y='test_score'  , hue ="gender" , style ="class_level" , size="subject")
```

```python
sns.relplot(kind ="scatter" , data=student , x='study_hours' , y='test_score'  , hue ="gender" , style ="class_level" )
```

```python
tips_data=sns.load_dataset("tips") # seaborn dataset 
tips_data.head()
```

```python
sns.relplot(kind ="scatter" , data =tips_data , x="total_bill" , y="tip" , hue="sex" , style="time" , size="size") 
```

```python
student1= pd.read_csv("./student_completed_data.csv")
student1.head()
```

Facet Plot

```python
#facet plot is used to create multiple subplots based 
sns.relplot(kind="scatter" , data =student1 , x="study_hours" , y="test_score" , hue="gender")
```

```python
sns.relplot(kind="scatter" , data =student1 , x="study_hours" , y="test_score" , hue="gender" ,col ="gender")
```

```python
sns.relplot(kind="scatter" , data =student1 , x="study_hours" , y="test_score" , col="gender" , row="class_name" )
```

```python
sns.relplot(kind="scatter" , data=student1 , x='study_hours' , y='test_score'  ,hue="gender" , col="week" , col_wrap=2)
```

Histogram using Seaborn

```python
sns.histplot(data =student1 , x="study_hours" ,hue="gender" , bins=10  )
```

```python
sns.displot(kind="hist" , data=student1 , x="test_score" , col="gender" ) 
```

```python
sns.displot(kind="hist" , data=tips_data , x="tip" , bins=20) # figure level histogram
```

KDE Plot using Seaborn
KDE  is a smoothed version of a histogram and provides a continuous probability density curve.

```python
sns.kdeplot(data=tips_data , x="tip" , hue="sex" )
```

```python
#figure level kde 
sns.displot(kind="kde" , data = tips_data , x="tip" , col="sex")
```

CountPlot using Seaborn

```python
student1.head() 
```

```python
sns.countplot(data =student1 , x="class_name"  , hue="gender")
```

Bar plot

```python

sns.barplot(data =student1, x="gender" , y ="test_score" , errorbar=None) 
#mean value of test score for each gender
```

```python
sns.barplot(data =student1, x="gender" , y ="test_score" , errorbar=None, estimator = np.median)

#median value of test score for each gender
```

```python
sns.barplot(data =student1, x="gender" , y ="test_score" , errorbar=None, estimator = np.max)

#max value of test score for each gender
```

RegPlot

```python
sns.regplot(data=student1 , x="study_hours" , y="test_score" ) 
#axes level plot for regression
```

```python
sns.lmplot(data=student1 , x="study_hours" , y="test_score" , hue="gender" )
#figure level plot for regression with hue
```
