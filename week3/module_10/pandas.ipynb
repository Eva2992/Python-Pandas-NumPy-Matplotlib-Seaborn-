DataFrame from Inbuilt 

```python
# list to dataFrame
mylist = [['Alice' , 85] ,
          ['Bob' , 90] ,
          ['Charlie' , 78 ]]  # 2D list 

list_dataFrame= pd.DataFrame(mylist , columns=['name' , 'marks']) # list to dataFrame
print(list_dataFrame)

```

```python
#tuple to dataFrame
mytuple= (('Alice' , 56) ,
          ('Bob' , 70) ,
          ('charlie' , 80)) # 2D tuple

tuple_dataFrame = pd.DataFrame(mytuple , columns=['name' , 'marks']) 
print(tuple_dataFrame)
```

```python
# dictionary to dataFrame

mydictionary= {
    'name' : ['Alice'   , 'Bob' , 'Charlie'] ,
    'marks' : [56 , 70 , 80]}

dict_dataFrame = pd.DataFrame(mydictionary)
print(dict_dataFrame)
```

```python
mydata = [
    {'name' : 'Alice' , 'marks' : 85} ,
    {'name' : 'Bob' , 'marks' : 90} ,
    {'name' : 'Charlie' }
]   # list of dictionary

dataFrame_from_list_of_dict = pd.DataFrame(mydata)
print(dataFrame_from_list_of_dict)
```
